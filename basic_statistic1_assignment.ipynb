# Standardization of Numerical Variables

●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.
                                                                                               
●	Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
	
	Show before and after comparisons of the data distributions.


```python
# Explanation
'''
Explanation of Standardization (Z-Score Normalization)
Standardization, also called z-score normalization, transforms numerical variables to have a mean of 0 and a standard deviation of 1.
This is especially useful for machine learning models that assume the data is normally distributed or sensitive to the scale of input variables.
The formula for standardization is:

𝑧 = (𝑥−𝜇)/𝜎

where:
x is the original value.
μ is the mean of the column,
σ is the standard deviation of the column.'''

#code
numerical_data = df.select_dtypes(include='number')
mean = numerical_data.mean()
std_dev = numerical_data.std()
standardized_data = (numerical_data - mean)/std_dev

for col in numerical_data.columns:
    plt.figure(figsize=(14,5))
    #plot original distribution
    plt.subplot(1,2,1)
    sns.histplot(numerical_data[col], kde=True, color='brown')
    plt.title(f"Original Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel('Frequency')

    #plot standardised distribution
    plt.subplot(1,2,2)
    sns.histplot(standardized_data[col], kde=True, color='skyblue')
    plt.title(f'standardized Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
    
```

# Conversion of Categorical Data into Dummy Variables

●	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
    
●	Steps:
	Discuss the need for converting categorical data into dummy variables (one-hot encoding).
	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
	Display a portion of the transformed dataset.


```python
''' Converting categorical data into dummy variables ensures compatibility with machine learning algorithms
    that cannot process textual data directly.
    It removes any implied ordinal relationship between categories and treats each category as an independent feature,
    improving model interpretability and accuracy.
'''

categorical_data = df.select_dtypes(include=['category','object'])
df_encoded = pd.get_dummies(categorical_data)
df_final = pd.concat([df.drop(columns=categorical_data.columns), df_encoded], axis=1)
df_final.head()

```
