### Association between Device Type and Customer Satisfaction

#### Background:

Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices:
Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the
type of device purchased and the customer's satisfaction level.

                                                                      
#### Data Provided:
                                                                      
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:
Satisfaction	Smart Thermostat	Smart Light	Total
Very Satisfied	50	70	120
Satisfied	80	100	180
Neutral	60	90	150
Unsatisfied	30	50	80
Very Unsatisfied	20	50	70
Total	240	360	600
                                                                                                        
                                                                                                        
Objective:
To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.
Assignment Tasks:
                                                                                                        
#### 1. State the Hypotheses:
                                                                                                        
#### 2. Compute the Chi-Square Statistic:
                                                                                                        
#### 3. Determine the Critical Value:
Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)
                                                                                                        
#### 4. Make a Decision:
Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.
Submission Guidelines:
•	Provide a detailed report of your analysis, including each step outlined in the assignment tasks in a python file.
•	Include all calculations, the Chi-Square statistic, the critical value, and your conclusion.


```python
import numpy as np
from scipy.stats import chi2_contingency, chi2

# Data Provided
observed = np.array([
    [50, 70],  # Very Satisfied
    [80, 100],  # Satisfied
    [60, 90],  # Neutral
    [30, 50],  # Unsatisfied
    [20, 50]   # Very Unsatisfied
])

# Step 1: State the Hypotheses
# H0: There is no significant association between the type of smart home device and customer satisfaction level.
# H1: There is a significant association between the type of smart home device and customer satisfaction level.

# Step 2: Compute the Chi-Square Statistic
chi2_stat, p_value, dof, expected = chi2_contingency(observed)

# Step 3: Determine the Critical Value
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, dof)

# Step 4: Make a Decision
reject_null = chi2_stat > critical_value

# Step 5: Conclusion
if reject_null:
    conclusion = "Reject the null hypothesis. There is a significant association between the type of smart home device and customer satisfaction level."
else:
    conclusion = "Fail to reject the null hypothesis. There is no significant association between the type of smart home device and customer satisfaction level."

# Print the results
print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
print(f"\nChi-Square Statistic: {chi2_stat:.2f}")
print(f"Degrees of Freedom: {dof}")
print(f"Critical Value: {critical_value:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"Decision: {'Reject H0' if reject_null else 'Fail to reject H0'}")
print(f"Conclusion: {conclusion}")
```
